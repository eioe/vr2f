```python
import os
from os import path as op
import pandas as pd
import json
```

```python
p_in = 'C:/Users/Felix/Downloads/'
fname = "StereoFEM_test01.json"
os.listdir('configFiles')
```

```python
with open(op.join(p_in, fname)) as f:
    data =  json.loads(f.read())
```

```python
cols = data["trialSequenceKeys"]
df = pd.json_normalize(data, record_path =['trialSequenceValues'])
df2 = pd.DataFrame(df["values"].to_list(), columns=cols)
df3 = df2.astype(dtype={"block_num":"int",
                        "block_type":"int",
                        "stereo":"int",
                        "avatar":"int",
                        "emo":"int",
                        "order":"string"})
df3 = df3.astype({'stereo':'bool'})
df3
#df2.dtypes
```
