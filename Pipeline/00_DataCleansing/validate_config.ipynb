```python
import os
from os import path as op
from pathlib import Path
import json
import pandas as pd
```

```python
sub_path = Path('..', 'Data', 'Subjects')
sub_list_str = ['VR2FEM_S21']  #  os.listdir(sub_path)
mega_df = pd.DataFrame(columns=['id'])
for subID in sub_list_str:
    p_data = Path('..', 'Data', 'Subjects', subID, 'MainTask', 'Unity', 'S001', 'session_info')
    fname = Path(p_data, 'settings.json')
    with open(fname, 'r') as f:
        sett = json.load(f)
```
